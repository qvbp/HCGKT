```python
# import pandas as pd
# import re

# file = './data/assist2009/skill_builder_data_corrected_collapsed.csv'
# df = pd.read_csv(file)

# cnt = 0

# def clean_number(num_str):
#     # 先检查是否为NaN
#     if pd.isna(num_str):
#         return None
    
#     # 转换为字符串
#     num_str = str(num_str)
    
#     # 如果是浮点数（有小数点），且小数点后都是0
#     if re.match(r'^\d+\.0+$', num_str):
#         return str(int(float(num_str)))
#     return num_str

# # 创建一个字典来存储结果
# result = {}
# set_skill = set()

# # 遍历csv文件
# for index, row in df.iterrows():
#     cnt += 1
#     # 获取技能id
#     skill_id = row['skill_id']
    
#     # 跳过NaN和"NA"
#     if pd.isna(skill_id) or skill_id == "NA":
#         continue
    
#     # 将skill_id转换为字符串并分割
#     skill_id = str(skill_id).split('_')
    
#     cleaned_skill_ids = []
#     for id_part in skill_id:
#         cleaned_id = clean_number(id_part)
#         if cleaned_id is not None:
#             cleaned_skill_ids.append(cleaned_id)
#             set_skill.add(cleaned_id)
#         if cleaned_id == '101':
#             print(f"cnt: {cnt}")
#             print(row['skill_name'])
#             print(row['skill_id'])
#             print(cleaned_id)
#     # 对id进行映射
#     for skill_id_part in cleaned_skill_ids:
#         skill_idx = skill_mapping.get(skill_id_part, None)
#         # if skill_id_part == '76':
#         #     print(skill_idx)
#         if skill_idx is not None:
#             result[skill_idx] = row['skill_name']

# # 使用 int() 转换键进行排序
# sorted_dict = dict(sorted(result.items(), key=lambda x: int(x[0])))
# print(sorted_dict)
# print(len(set_skill))
# print(set_skill)


import pandas as pd
import re
import random
import string

# 添加生成随机字符串的函数
def generate_random_string(length=10):
    """生成指定长度的随机字符串"""
    letters = string.ascii_letters  # 包含所有ASCII字母
    return ''.join(random.choice(letters) for _ in range(length))

file = './data/assist2009/skill_builder_data_corrected_collapsed.csv'
df = pd.read_csv(file)

cnt = 0

def clean_number(num_str):
    if pd.isna(num_str):
        return None
    
    num_str = str(num_str)
    
    if re.match(r'^\d+\.0+$', num_str):
        return str(int(float(num_str)))
    return num_str

# 创建一个字典来存储结果
result = {}
set_skill = set()

# 遍历csv文件
for index, row in df.iterrows():
    cnt += 1
    skill_id = row['skill_id']
    
    if pd.isna(skill_id) or skill_id == "NA":
        continue
    
    skill_id = str(skill_id).split('_')
    
    cleaned_skill_ids = []
    for id_part in skill_id:
        cleaned_id = clean_number(id_part)
        if cleaned_id is not None:
            cleaned_skill_ids.append(cleaned_id)
            set_skill.add(cleaned_id)
        # if cleaned_id == '101':
        #     print(f"cnt: {cnt}")
        #     print(row['skill_name'])
        #     print(row['skill_id'])
        #     print(cleaned_id)

    for skill_id_part in cleaned_skill_ids:
        skill_idx = skill_mapping.get(skill_id_part, None)
        if skill_idx is not None:
            result[skill_idx] = row['skill_name']

# 对结果进行处理：拼接key值并处理nan
final_result = {}
for key, value in sorted(result.items(), key=lambda x: int(x[0])):
    if pd.isna(value) or value == 'nan':
        # 如果值为nan，生成随机字符串
        new_value = f"{generate_random_string()}_{key}"
    else:
        # 否则直接拼接key
        new_value = f"{value}_{key}"
    final_result[key] = new_value

# 使用 int() 转换键进行排序并打印
sorted_final_dict = dict(sorted(final_result.items(), key=lambda x: int(x[0])))

print(sorted_final_dict)
print(len(set_skill))
print(set_skill)
```

```python
# 使用 json.dump() - 直接写入文件
with open('./KC_Text_change/data/kcs_context.json', 'w', encoding='utf-8') as f:
    json.dump(sorted_final_dict, f, ensure_ascii=False, indent=4)
```

```python
file = './data/assist2009/skill_builder_data_corrected_collapsed.csv'
df = pd.read_csv(file)

# 数据清理：替换空字符串和 'NA'，确保统一格式
df['skill_name'] = (
    df['skill_name']
    .astype(str)                   # 确保所有值是字符串
    .replace(['', 'NA', 'nan'], None)  # 替换无效值为 None
    .fillna('Missing')             # 填充占位符，方便后续操作
    .str.strip()                   # 去掉前后空格
)

df['skill_id'] = (
    df['skill_id']
    .astype(str)                   # 确保所有值是字符串
    .replace(['', 'NA', 'nan'], None)  # 替换无效值为 None
    .fillna('Missing')             # 填充占位符，方便后续操作
    .str.strip()                   # 去掉前后空格
)

# 读取csv中其中一列的内容转成列表并去重
kc_content_name = [x for x in df['skill_name'] if x != 'Missing']
kc_content_id = [x for x in df['skill_id'] if x != 'Missing']
# print(kc_content_name[:10])
# print(kc_content_id[:10])

# 对列表中的内容去重
kc_content_name_clean = list(set(kc_content_name))
kc_content_id_clean = list(set(kc_content_id))
print(kc_content_name_clean)
print(kc_content_id_clean)
print(len(kc_content_name_clean))
print(len(kc_content_id_clean))
```

```python
# 创建一个字典来存储结果
result = {}

# 搞一个字典映射
for i in range(len(kc_content_id_clean)):
    nums = kc_content_id_clean[i].split('_')
    for num in nums:
        result[num] = kc_content_name_clean[i]
print(len(result))
print(result[:10])
print(result[-10:])
```

# al2005 概念文本json文件获取

```python
# 读取json文件，获取里面的映射字典
with open('./data/algebra2005/keyid2idx.json', 'r') as f:
    data = json.load(f)

# 获取映射字典
skill_mapping = data['concepts']

# 创建一个字典来存储结果
result = {}

# skill_mapping这个字典的key是知识点文本，value是知识点id，现在需要把它们反过来，然后存到result里，并且需要修改key里面文本中所有"@@@@"为","，所有的"####"为"_"
for key, value in skill_mapping.items():
    new_key = key.replace('@@@@', ',').replace('####', '_')
    result[value] = new_key

# 保存结果
with open('./KC_Text_change/data/kcs_context_algebra2005.json', 'w', encoding='utf-8') as f:
    json.dump(result, f, ensure_ascii=False, indent=4)

print("Done!")

```

# nips_task34概念文本json文件获取

```python
# 读取json文件，获取里面的映射字典
with open('./data/nips_task34/keyid2idx.json', 'r') as f:
    data = json.load(f)

# 获取映射字典
skill_mapping = data['concepts']

# 创建一个字典来存储结果
result = {}

# 读取csv文件
df = pd.read_csv('./data/nips_task34/metadata/subject_metadata.csv')

# 遍历csv文件每一行，然后判断“SubjectId”是否在skill_mapping中，如果在就把skill_mapping中的value作为key，然后把“Name”作为value存到result中，但要先把“SubjectId”转为字符串
for index, row in df.iterrows():
    subject_id = str(row['SubjectId'])
    print(subject_id)
    if subject_id in skill_mapping:
        result[skill_mapping[subject_id]] = row['Name']

# 输出一下长度看一下是否正确
print(len(result))

# 将result按照key升序排序
result = dict(sorted(result.items(), key=lambda x: int(x[0])))

# 保存结果
with open('./KC_Text_change/data/kcs_context_nips_task34.json', 'w', encoding='utf-8') as f:
    json.dump(result, f, ensure_ascii=False, indent=4)

print("Done!")

```
