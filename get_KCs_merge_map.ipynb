# 统计问题准确率

```python
# Read the data from the CSV file
train_df = pd.read_csv('./data/algebra2005/train_valid_quelevel.csv')
test_df = pd.read_csv('./data/algebra2005/test_quelevel.csv')

# define number of questions
num_questions = 17737
# define right and wrong answer list
right_answer_list = [0 for i in range(num_questions)]
wrong_answer_list = [0 for i in range(num_questions)]

# itterate over the train data
for index, row in train_df.iterrows():
    # 获取question_id和response
    question_id = [int(x) for x in row['questions'].split(',')]
    response = [int(x) for x in row['responses'].split(',')]

    # 开始按questions填充正确和错误列表
    for i in range(len(question_id)):
        if response[i] == 1:
            right_answer_list[question_id[i]] += 1
        else:
            wrong_answer_list[question_id[i]] += 1

# # itterate over the test data
# for index, row in test_df.iterrows():
#     # 获取question_id和response
#     question_id = [int(x) for x in row['questions'].split(',')]
#     response = [int(x) for x in row['responses'].split(',')]

#     # 开始按questions填充正确和错误列表
#     for i in range(len(question_id)):
#         if response[i] == 1:
#             right_answer_list[question_id[i]] += 1
#         else:
#             wrong_answer_list[question_id[i]] += 1

# calculate the correction of each question
question_correction = []
for i in range(num_questions):
    if right_answer_list[i] + wrong_answer_list[i] == 0:
        question_correction.append(0)
    else:
        question_correction.append(right_answer_list[i] / (right_answer_list[i] + wrong_answer_list[i]))

# save the correction to a csv file
pd.DataFrame(question_correction).to_csv('./data/algebra2005/question_correction_train_valid.csv', index=False, header='correction')
# print done
print("Done")
```

# 统计KC准确率

```python
# Read the data from the CSV file
train_df = pd.read_csv('./data/algebra2005/train_valid_quelevel.csv')
test_df = pd.read_csv('./data/algebra2005/test_quelevel.csv')

# define number of questions
num_concepts = 123
# define right and wrong answer list
right_answer_list = [0 for i in range(num_concepts)]
wrong_answer_list = [0 for i in range(num_concepts)]

# itterate over the train data
for index, row in train_df.iterrows():
    # 获取concept_id和response
    response = [int(x) for x in row['responses'].split(',')]

    concept_id = row['concepts'].split(',')
    # print(concept_id)
    for i in range(len(concept_id)):
        kc = [int(x) for x in concept_id[i].split('_')]
        for j in range(len(kc)):
            if response[i] == 1:
                right_answer_list[kc[j]] += 1
            else:
                wrong_answer_list[kc[j]] += 1

# # itterate over the train data
# for index, row in test_df.iterrows():
#     # 获取concept_id和response
#     response = [int(x) for x in row['responses'].split(',')]

#     concept_id = row['concepts'].split(',')
#     for i in range(len(concept_id)):
#         kc = [int(x) for x in concept_id[i].split('_')]
#         for j in range(len(kc)):
#             if response[i] == 1:
#                 right_answer_list[kc[j]] += 1
#             else:
#                 wrong_answer_list[kc[j]] += 1

# calculate the correction of each question
concept_correction = []
for i in range(num_concepts):
    if right_answer_list[i] + wrong_answer_list[i] == 0:
        concept_correction.append(0)
    else:
        concept_correction.append(right_answer_list[i] / (right_answer_list[i] + wrong_answer_list[i]))

# save the correction to a csv file
pd.DataFrame(concept_correction).to_csv('./data/algebra2005/concept_correction_train_valid.csv', index=False, header='correction')
# print done
print("Done")
```

# 统计问题的出现次数

```python
# Read the data from the CSV file
train_df = pd.read_csv('./data/algebra2005/train_valid_quelevel.csv')
test_df = pd.read_csv('./data/algebra2005/test_quelevel.csv')

# define number of questions
num_questions = 17737
# define right and wrong answer list
num_questions_list_train_valid = [0 for i in range(num_questions)]
num_questions_list_test = [0 for i in range(num_questions)]
num_questions_list_all = [0 for i in range(num_questions)]

# itterate over the train data
for index, row in train_df.iterrows():
    # 获取question_id和response
    question_id = [int(x) for x in row['questions'].split(',')]
    for i in range(len(question_id)):
        num_questions_list_train_valid[question_id[i]] += 1

# itterate over the test data
for index, row in test_df.iterrows():
    # 获取question_id和response
    question_id = [int(x) for x in row['questions'].split(',')]
    for i in range(len(question_id)):
        num_questions_list_test[question_id[i]] += 1

# statistics all the question numbers
for i in range(num_questions):
    num_questions_list_all[i] = num_questions_list_train_valid[i] + num_questions_list_test[i]



# save the correction to a csv file
pd.DataFrame(num_questions_list_train_valid).to_csv('./data/algebra2005/question_counts_train_valid.csv', index=False, header='correction')
pd.DataFrame(num_questions_list_test).to_csv('./data/algebra2005/question_counts_test.csv', index=False, header='correction')
pd.DataFrame(num_questions_list_all).to_csv('./data/algebra2005/question_counts_all.csv', index=False, header='correction')

# print done
print("Done")
```

# 统计问题的KC，然后取平均

```python
import pandas as pd
import numpy as np
import json

# 读取 CSV 文件
data = pd.read_csv('./data/peiyou/ques_skills.csv') 

# 读取json文件，对问题编号和知识点编号做一个映射
with open('./data/peiyou/keyid2idx.json', 'r') as f:
    map = json.load(f)

kc_map = map['concepts']
ques_map = map['questions']

# 把csv中的问题编号和知识点编号映射到新的编号,但是json中的映射编号是str类型，而csv中的是int类型，所以要先转换一下
data['ques'] = data['ques'].apply(lambda x: ques_map[str(x)])
data['skill'] = data['skill'].apply(lambda x: kc_map[str(x)])

# 输出一下映射后的结果
print(data)

# 获取问题和知识点的最大编号
max_pro = data['ques'].max() + 1  # 问题编号从0开始，因此加1作为最大编号
grouped = data.groupby('ques')['skill'].apply(list)  # 分组后将关联的知识点收集成列表

# 获取一个问题最大关联的知识点数量
max_concepts = grouped.apply(len).max()

# 输出一下结果
print(max_pro, max_concepts)

# 初始化张量
tensor = np.full((max_pro, max_concepts), -1, dtype=int)

# 填充张量
for ques, skills in grouped.items():
    tensor[ques, :len(skills)] = skills

# 打印结果
print(tensor)
print(tensor.shape)

# 保存张量
np.save('./data/peiyou/question_concept_map.npy', tensor)
```
